```python
import numpy as np
from netCDF4 import Dataset, num2date
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import colors
import matplotlib.lines as mlines
import cartopy
from cartopy.io import shapereader
from cartopy.feature import ShapelyFeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import geopandas
import fiona
import shapely.geometry as sgeom
import shapely.vectorized as v
from shapely.prepared import prep
from shapely.ops import unary_union
from scipy.interpolate import interp2d
from scipy.ndimage import gaussian_filter as gf
```

```python
# we'll need a single, complete polygon to describe where land is
# we can do that by extracting the NE polygons and combining them, making sure there are no overlapping points etc

land_shapes = []

with fiona.open(
		shapereader.natural_earth(resolution='50m',
								  category='physical',
								  name='land')) as src:
	for feat in src:
		geom = sgeom.shape(feat['geometry'])
		if geom.geom_type == 'Polygon':
			land_shapes.append(geom)
		
		elif geom.geom_type == 'MultiPolygon':
			land_shapes.extend(geom.geoms)
		else:
			continue

land_poly = unary_union(land_shapes)
land      = prep(land_poly)
```

```python
# helper function to tell us if a given lon-lat point is over land or not

def is_land(x, y):
    return land.contains(sgeom.Point(x, y))
```

```python
def make_landmask(lons, lats):
	X, Y = np.meshgrid(lons, lats)
	return v.contains(land_poly, X, Y)	
```

```python
# compute the distance from the coast. We use this to determine possible offshore wind placement
# obviously better bathymetric methods exist, but this is illustrative

def dist_from_coast(lons, lats):
	X, Y = np.meshgrid(lons, lats)
	coast_distance = np.empty(X.shape, dtype=np.float64)
	for i, x_pt in enumerate(lons):
		for j, y_pt in enumerate(lats):
			coast_distance[j, i] = land_poly.distance(sgeom.Point(x_pt, y_pt))
	coast_distance *= 111*np.cos(Y*np.pi/180.)
	return(coast_distance)
```

```python
#function to compute the geodetic distance between two points

def haversine(lon1_, lat1_, lon2_, lat2_):
	lon1=lon1_*np.pi/180.
	lat1=lat1_*np.pi/180.
	lon2=lon2_*np.pi/180.
	lat2=lat2_*np.pi/180.
	dlon = lon2 - lon1 
	dlat = lat2 - lat1 
	a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
	c = 2 * np.arcsin(np.sqrt(a)) 
	r = 6371.
	return c * r 
```

```python
%matplotlib inline
plt.ioff()

fig = plt.figure(figsize=(8,8))
ax = plt.subplot(1,1,1, projection=cartopy.crs.PlateCarree())
xmin, xmax = 65, 100
ymin, ymax = 5, 40
```

```python
# load the respective capacity factor files plus others we'll need later

oro = np.load("../other-files/era5_orography.npy")

wind = np.load("../other-files/windcapacity.npy")
solar = np.load("../other-files/solarcapacity.npy")
waves = np.load("../other-files/wavemean.npy")
hydro = np.load("../other-files/hydropower.npy")

lulc = np.load("../other-files/lulc.npy")
trees = np.load("../other-files/treecoverage.npy")
```

```python
#compute "optimal" locations for onshore and offshore wind
# for offshore, require that mean capacity factor is at least in the 80th percentile
# and locations are within 250 km of the coast

# for onshore, require that the altitude is less than 2 km, and that the mean cap fac
# is in at least the 85th percentile

lons = np.linspace(xmin,xmax,wind.shape[1])
lats = np.linspace(ymax,ymin,wind.shape[0])

land_mask = make_landmask(lons, lats)
coast_distance = dist_from_coast(lons, lats)

offwind = wind*(1-land_mask)*(coast_distance<250)
thresh = np.percentile(offwind[offwind>0], 80)
print(thresh)

cs = ax.contourf(lons,lats, offwind, levels = [thresh,1e10], hatches=['\\\\\\',], colors='none', zorder=2)
for i, collection in enumerate(cs.collections):
	collection.set_edgecolor('tab:blue')
	

onwind = wind*land_mask*(oro<2000)
thresh = np.percentile(onwind[onwind>0], 85)
print(thresh)

cs = ax.contourf(lons,lats, onwind, levels = [thresh,1e10], hatches=['\\\\\\',], colors='none',zorder=2)
for i, collection in enumerate(cs.collections):
	collection.set_edgecolor('tab:red')
```

```python
# compute "optimal" locations for solar
# for conventional solar, require altitude less than 3 km, cap fac > 70th percentile
# for "high-altitude solar (i.e. above 3 km), specify that mean capacity factor must be above 90th %ile


isolar = solar*land_mask*(oro<3000)
thresh = np.percentile(isolar[isolar>0], 70)
print(thresh)
cs = ax.contourf(lons,lats, isolar, levels = [thresh,1e10], colors='yellow', zorder=1)

isolar = solar*land_mask*(oro>3000)
thresh = np.percentile(isolar[isolar>0], 90)
print(thresh)
cs = ax.contourf(lons,lats, isolar, levels = [thresh,1e10], colors='orange', zorder=1)
```

```python
# for wave power, identify gridpoints where mean wave power > 75%ile

lons = np.linspace(xmin,xmax,waves.shape[1])
lats = np.linspace(ymax,ymin,waves.shape[0])

thresh = np.percentile(waves[waves>0], 75)
print(thresh)
cs = ax.contourf(lons,lats, waves, levels = [thresh,1e10], colors='gray', zorder=1)
```

```python
# biomass: land type must be crop, grass, shrub, or herbaceous with at least 
# 50% tree cover in at least least one neighbouring 500 m gridbox
# this is obviously arbitrary, but shows where trees could be grown for this purpose

lons = np.linspace(xmin,xmax,lulc.shape[1])
lats = np.linspace(ymax,ymin,lulc.shape[0])

tlons = np.linspace(xmin,xmax,trees.shape[1])
tlats = np.linspace(ymax,ymin,trees.shape[0])

interp_trees = interp2d(tlons, tlats, trees)(lons, lats)
valid = gf(np.in1d(lulc, [2,3,5,6]).reshape((len(lats),len(lons))),(1,1))
good_land = interp_trees*valid

thresh = np.percentile(good_land[good_land>0], 90)
print(thresh)

cs = ax.contourf(lons,lats, good_land, levels = [thresh,1e10], colors='forestgreen', zorder=1)
```
