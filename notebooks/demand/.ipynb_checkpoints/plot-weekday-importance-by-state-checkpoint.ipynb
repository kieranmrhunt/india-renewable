```python
import numpy as np
from scipy.stats import linregress
from sklearn.metrics import mean_squared_error
import xgboost as xgb
import pandas as pd
import sys
import shap
from functions import *
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
# Load hyperparameters
hyperparameters_df = pd.read_csv("outputs/best_hyperparameters.csv")
hyperparameters_dict = hyperparameters_df.set_index('state').T.to_dict()

results_dict = {"State":[], "R Value":[], "Number of Trees":[]}

feature_names = feature_names_full
```

```python
%matplotlib inline
plt.ioff()

fig, ax_dict = plot_states_by_grid()
kerala_legend_ax = fig.add_axes([.7, .1, .2, .2])

day_mapping = {
    1: "Mon",
    2: "Tue",
    3: "Wed",
    4: "Thu",
    5: "Fri",
    6: "Sat",
    7: "Sun"
}
```

```python
for state, subdiv in states_subdiv_mapping.items():
	
	df, df_energy = load_data(state)
	df = process_data(df, df_energy, subdiv)
	df = df[feature_names + ['date', 'energy_met_MU']]
	df = df.dropna()
	
	ax_name = state.replace("_", " ").replace("and ", "& ")
	
	X = df[feature_names]
	y = df['energy_met_MU']
	
	best_params = hyperparameters_dict[state]
	
	xg_reg, X_test, y_test, preds = train_model(X, y, df, best_params)#, year=2019)
	r_value, n_trees = evaluate_model(xg_reg, X_test, y_test, preds, state=state)

	explainer = shap.Explainer(xg_reg)
	shap_values = explainer(X)
	
	idx = feature_names.index("day_of_week")
	idy = feature_names.index("is_holiday")
	
	base = y.std()
	data = shap_values.data[:, idx]
	hols = shap_values.data[:, idy]
	vals = shap_values.values[:, idx]/base
	
	shap_mean = np.array([np.mean(vals[data==i]) for i in range(1,8)])
	shap_25 = np.array([np.percentile(vals[data==i], 5) for i in range(1,8)])
	shap_75 = np.array([np.percentile(vals[data==i], 95) for i in range(1,8)])
	
	shap_holiday_mean = np.array([np.mean(vals[(data==i) & (hols==1)]) for i in range(1,8)])
	
	print(shap_mean)
	print(shap_holiday_mean)
	
	x = np.arange(1, 8)

	lower_error = shap_mean - shap_25
	upper_error = shap_75 - shap_mean
	asymmetric_error = [lower_error, upper_error]

	try:
		ax = ax_dict[ax_name]
	except: continue
	ax.errorbar(x, shap_mean, yerr=asymmetric_error, fmt='o', markersize=3, capsize=5, color='dodgerblue')
	ax.plot(x, shap_holiday_mean, 'kx', zorder=5)
	ax.set_ylim([-0.75,0.2])
	ax.set_xticks([])
	ax.set_xticklabels([])
	ax.set_yticks(np.arange(-0.6, 0.4, 0.2))
	ax.set_yticklabels([])
	ax.text(0.5, 0.1, subdiv, ha='center', va='center', transform=ax.transAxes)

	if r_value**2 < 0.5:
		ax.patch.set_facecolor('grey')
		ax.patch.set_alpha(0.5)

	if state == "Kerala":
		
		kerala_legend_ax.errorbar(x, shap_mean, yerr=asymmetric_error, fmt='o', markersize=3, capsize=5, color='dodgerblue', label = "All days")
		kerala_legend_ax.plot(x, shap_holiday_mean, 'kx', zorder=5, label='Holidays')
		kerala_legend_ax.set_title("Kerala (legend)")
		kerala_legend_ax.set_xticks(list(day_mapping.keys()))
		kerala_legend_ax.set_xticklabels(list(day_mapping.values()))
		kerala_legend_ax.set_xlabel("Day of week")
		kerala_legend_ax.set_ylabel("Impact on energy demand\n(standardised Shapley values)")
		kerala_legend_ax.set_ylim([-0.75,0.2])
		plt.legend(loc='lower left')
	
	
plt.show()
```
