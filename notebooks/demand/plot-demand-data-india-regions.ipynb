```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
import numpy as np
import cartopy.crs as ccrs
import geopandas as gpd
import matplotlib.gridspec as gridspec
%matplotlib inline
plt.ioff()
```

```python
# Define the zoom date range
start_zoom = datetime(2023, 1, 1)
end_zoom = datetime(2024, 1, 1)
```

```python
# Create a gridspec with width ratios to adjust subplot widths
fig = plt.figure(figsize=(15, 10))
gs = gridspec.GridSpec(2, 2, width_ratios=[2, 1])  # Left plots twice as wide as right plots

ax1 = fig.add_subplot(gs[0, 0])
ax1_zoom = fig.add_subplot(gs[0, 1]) 
ax2 = fig.add_subplot(gs[1, 0])
ax2_zoom = fig.add_subplot(gs[1, 1]) 
```

```python
df = pd.read_csv("outputs/corrected_daily_energy_met_MU.csv")
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#Define color keys and regions
color_keys = {'NER': 'lightgrey', 'ER': 'gold', 'SR': 'tab:blue',
              'WR': 'indianred', 'NR': 'limegreen', 'I': 'w'}

regions = {  
    'Assam': 'NER',
    'Dadra and Nagar Haveli': 'WR',
    'Maharashtra': 'WR',
    'Madhya Pradesh': 'WR',
    'Arunachal Pradesh': 'NER',
    'Bihar': 'ER',
    'Tripura': 'NER',
    'Lakshadweep': 'I',
    'Haryana': 'NR',
    'Daman and Diu': 'WR',
    'Delhi': 'NR',
    'Meghalaya': 'NER',
    'Andhra Pradesh': 'SR',
    'Sikkim': 'ER',
    'Manipur': 'NER',
    'Rajasthan': 'NR',
    'Kerala': 'SR',
    'Nagaland': 'NER',
    'Jammu and Kashmir': 'NR',
    'Jharkhand': 'ER',
    'Tamil Nadu': 'SR',
    'Gujarat': 'WR',
    'Uttar Pradesh': 'NR',
    'Karnataka': 'SR',
    'Ladakh': 'NR',
    'Mizoram': 'NER',
    'Andaman and Nicobar Islands': 'I',
    'Goa': 'WR',
    'Himachal Pradesh': 'NR',
    'West Bengal': 'ER',
    'Telangana': 'SR',
    'Chandigarh': 'NR',
    'Uttarakhand': 'NR',
    'Chhattisgarh': 'WR',
    'Odisha': 'ER',
    'Punjab': 'NR',
    'Puducherry': 'SR',
    'AMNSIL': 'I',
    'Damodar Valley Corporation': 'I'
}
```

```python
# Prepare totals per region
totals = {'NR': [], 'ER': [], 'SR': [], 'WR': [], 'I': [], 'NER': []}

for state in df.columns[1:-1]:
    region = regions[state]
    totals[region].append(df[state])

total_region = {}

for region in ['NR', 'ER', 'WR', 'NER', 'SR']:
    total_region[region] = np.nansum(totals[region], axis=0)
```

```python
# Plot on ax1
for region in ['NR', 'ER', 'WR', 'NER', 'SR']:
    ax1.plot(df['Date'], total_region[region],
             color=color_keys[region], label=region)

ax1.set_xlim([datetime(2013, 1, 1), datetime(2024, 1, 1)])
ax1.set_ylabel("Total daily energy demand (MU)")
ax1.legend(loc='upper left')
```

```python
# Create zoomed-in date mask
mask_zoom = (df['Date'] >= start_zoom) & (df['Date'] < end_zoom)

# Plot on ax1_zoom
for region in ['NR', 'ER', 'WR', 'NER', 'SR']:
    ax1_zoom.plot(df['Date'][mask_zoom], total_region[region][mask_zoom],
                  color=color_keys[region])

ax1_zoom.set_xlim([start_zoom, end_zoom])
ax1_zoom.set_ylim(ax1.get_ylim())  # Match y-limits with parent
ax1_zoom.set_ylabel('')  # Remove y-axis label

# Rotate date labels on ax1_zoom
for label in ax1_zoom.get_xticklabels():
    label.set_rotation(35)
    label.set_horizontalalignment('right')
```

```python
### Second subplot

case_state = "Uttar Pradesh"
state_energy = df[case_state]
state_filename = case_state.replace(" ", "_")

state_df = pd.read_csv(f"../../demand/regression/data/daily/{state_filename}.csv")
state_df['date'] = pd.to_datetime(state_df['date'])
state_df['mean_wind_speed'] = np.sqrt(
    state_df['10m_u_component_of_wind_mean']**2 +
    state_df['10m_v_component_of_wind_mean']**2)
state_df['2m_temperature_mean'] = state_df['2m_temperature_mean'] - 273.15
state_df['2m_dewpoint_temperature_max'] = state_df['2m_dewpoint_temperature_max'] - 273.15
```

```python
# Create date masks for zoom
mask_zoom_df = (df['Date'] >= start_zoom) & (df['Date'] < end_zoom)
mask_zoom_state = (state_df['date'] >= start_zoom) & (state_df['date'] < end_zoom)

# Plot on ax2
host = ax2
host.plot(df['Date'], state_energy, 'k-', lw=2, zorder=10)
ax2.text(0.025, 0.98, case_state, transform=ax2.transAxes, ha='left', va='top')

# Set y-axis label
host.set_ylabel('Daily energy demand (MU)')

# Make ax2 background transparent
host.set_facecolor('none')
```

```python
# Create parasite axes
temp_axis = host.twinx()
dewpoint_axis = host.twinx()
wind_speed_axis = host.twinx()

# Offset the extra axes
dewpoint_axis.spines['right'].set_position(('outward', 60))
wind_speed_axis.spines['right'].set_position(('outward', 120))

# Plot the additional data
temp_axis.plot(state_df['date'], state_df['2m_temperature_mean'],
               'r-', lw=0.5)
dewpoint_axis.plot(state_df['date'], state_df['2m_dewpoint_temperature_max'],
                   'b-', lw=0.5)
wind_speed_axis.plot(state_df['date'], state_df['mean_wind_speed'],
                     'lightgrey', lw=0.5)

# Adjust z-order and frame
host.set_zorder(10)
host.set_frame_on(True)  # Keep frame on to show the right spine
```

```python
# Hide ticks and tick labels on parasite axes for ax2
for axis in [temp_axis, dewpoint_axis, wind_speed_axis]:
    axis.tick_params(axis='y', which='both', left=False, right=False,
                     labelleft=False, labelright=False)
    axis.set_ylabel('')  # Remove y-axis labels
    axis.spines['right'].set_visible(False)  # Hide spines

host.set_xlim([datetime(2013, 1, 1), datetime(2024, 1, 1)])

host.spines['right'].set_visible(True)  # Show the right spine
```

```python
# Adjust the y-axis limits to remove separation
separation_parameter = 1  # Set separator to zero
host.set_ylim(host.get_ylim())  
temp_axis.set_ylim(temp_axis.get_ylim()) 
dewpoint_axis.set_ylim(dewpoint_axis.get_ylim()) 
wind_speed_axis.set_ylim(wind_speed_axis.get_ylim()) 
```

```python
# Plot on ax2_zoom
host_zoom = ax2_zoom
host_zoom.plot(df['Date'][mask_zoom_df], state_energy[mask_zoom_df],
               'k-', lw=2, zorder=10)
ax2_zoom.text(0.025, 0.98, case_state, transform=ax2_zoom.transAxes,
              ha='left', va='top')
host_zoom.set_ylim(host.get_ylim())  # Match y-limits with parent
host_zoom.set_ylabel('')  # Remove y-axis label
```

```python
# Create parasite axes for zoom
temp_axis_zoom = host_zoom.twinx()
dewpoint_axis_zoom = host_zoom.twinx()
wind_speed_axis_zoom = host_zoom.twinx()

# Offset the extra axes
dewpoint_axis_zoom.spines['right'].set_position(('outward', 60))
wind_speed_axis_zoom.spines['right'].set_position(('outward', 120))

# Plot the additional data
temp_axis_zoom.plot(state_df['date'][mask_zoom_state],
                    state_df['2m_temperature_mean'][mask_zoom_state],
                    'r-', lw=0.5)
dewpoint_axis_zoom.plot(state_df['date'][mask_zoom_state],
                        state_df['2m_dewpoint_temperature_max'][mask_zoom_state],
                        'b-', lw=0.5)
wind_speed_axis_zoom.plot(state_df['date'][mask_zoom_state],
                          state_df['mean_wind_speed'][mask_zoom_state],
                          'lightgrey', lw=0.5)
```

```python
# Adjust z-order and frame
host_zoom.set_zorder(10)
host_zoom.set_frame_on(False)

# Set axis labels and colors
host_zoom.set_xlabel('Date')
temp_axis_zoom.set_ylabel('2-m daily mean temperature (°C)', color='r')
dewpoint_axis_zoom.set_ylabel('2-m daily max dewpoint (°C)', color='b')
wind_speed_axis_zoom.set_ylabel('10-m daily mean wind speed (m s$^{-1}$)', color='grey')

temp_axis_zoom.yaxis.label.set_color('red')
dewpoint_axis_zoom.yaxis.label.set_color('blue')
wind_speed_axis_zoom.yaxis.label.set_color('grey')

temp_axis_zoom.tick_params(axis='y', colors='red')
dewpoint_axis_zoom.tick_params(axis='y', colors='blue')
wind_speed_axis_zoom.tick_params(axis='y', colors='grey')

host_zoom.set_xlim([start_zoom, end_zoom])

# Set y-limits of parasite axes to match parent axes
temp_axis_zoom.set_ylim(temp_axis.get_ylim())
dewpoint_axis_zoom.set_ylim(dewpoint_axis.get_ylim())
wind_speed_axis_zoom.set_ylim(wind_speed_axis.get_ylim())
```

```python
# Rotate date labels on ax2_zoom
for label in host_zoom.get_xticklabels():
    label.set_rotation(35)
    label.set_horizontalalignment('right')

# Adjust the position of the inset map and make it bigger
ax3 = fig.add_axes([0.165, 0.75, 0.125, 0.125], projection=ccrs.PlateCarree())

states_df = gpd.read_file("../other-files/india_states_fixed.shp")

for state in regions:
    gf = states_df[states_df.ST_NM == state]
    try:
        gf.plot(ax=ax3, color=color_keys[regions[state]],
                edgecolor='grey', linewidth=0)
    except:
        pass

# Add map details
ax3.set_extent([68, 98, 6, 38])  # Set the geographical extent of India
ax3.axis('off')  # Turn off the axis

plt.show()
```
